## step 1 : Identifying Input and Output

```python
y = df.pop('loan_status')
X = df
```

## step 2 : Test Train Split

```python
# Split into train and test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.25, 
                                                    random_state = 0)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# Separating Categorical and Numerical Columns

X_train_cat = X_train.select_dtypes(include=['object'])
X_train_num = X_train.select_dtypes(include=['int64', 'float64'])
```

```python
person_home_ownership_encoder = {'RENT' : 1, 'MORTGAGE' : 2,'OWN' : 3,'OTHER' : 4}
loan_intent_encoder = {'EDUCATION':1, 'MEDICAL':2, 'VENTURE':3, 'PERSONAL':4, 'DEBTCONSOLIDATION':5,'HOMEIMPROVEMENT':6}
loan_grade_encoder = {'A':1, 'B':2,'C':3,'D':4,'E':5,'F':6,'G':7}
cb_person_default_on_file_encoder = {'Y':1, 'N':2}




# Apply Label Encoding
X_train_cat['person_home_ownership'] = X_train_cat['person_home_ownership'].map(person_home_ownership_encoder)
X_train_cat['loan_intent'] = X_train_cat['loan_intent'].map(loan_intent_encoder)
X_train_cat['loan_grade'] = X_train_cat['loan_grade'].map(loan_grade_encoder)
X_train_cat['cb_person_default_on_file'] = X_train_cat['cb_person_default_on_file'].map(cb_person_default_on_file_encoder)

```

```python
X_train_cat
```

```python
X_train_transformed = pd.concat([X_train_num,X_train_cat], axis = 1)
```

```python
X_train_transformed.head()
```

```python
from sklearn.tree import DecisionTreeClassifier 

# create a regressor object 
classifier = DecisionTreeClassifier() 

# fit the regressor with X and Y data 
classifier.fit(X_train_transformed, y_train) 
```

```python
classifier.feature_importances_
```

```python
# Rescaling numerical features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# column names are (annoyingly) lost after Scaling
# (i.e. the dataframe is converted to a numpy ndarray)
X_train_num_transformed = pd.DataFrame(scaler.fit_transform(X_train_num), 
                                    columns = scaler.get_feature_names_out(), 
                                    index = X_train_num.index)

X_train_num_transformed.head()
```

```python
X_train_transformed = pd.concat([X_train_num_transformed,X_train_cat], axis = 1)
```

```python
X_train_transformed.head()
```

```python
X_test_cat = X_test.select_dtypes(include=['object'])
X_test_num = X_test.select_dtypes(include=['int64', 'float64'])
```

```python
X_test_num_transformed = pd.DataFrame(scaler.transform(X_test_num), 
                                   columns = scaler.get_feature_names_out(), 
                                   index = X_test_num.index)

X_test_num_transformed.head()
```

```python
person_home_ownership_encoder = {'RENT' : 1, 'MORTGAGE' : 2,'OWN' : 3,'OTHER' : 4}
loan_intent_encoder = {'EDUCATION':1, 'MEDICAL':2, 'VENTURE':3, 'PERSONAL':4, 'DEBTCONSOLIDATION':5,'HOMEIMPROVEMENT':6}
loan_grade_encoder = {'A':1, 'B':2,'C':3,'D':4,'E':5,'F':6,'G':7}
cb_person_default_on_file_encoder = {'Y':1, 'N':2}




# Apply Label Encoding
X_test_cat['person_home_ownership'] = X_test_cat['person_home_ownership'].map(person_home_ownership_encoder)
X_test_cat['loan_intent'] = X_test_cat['loan_intent'].map(loan_intent_encoder)
X_test_cat['loan_grade'] = X_test_cat['loan_grade'].map(loan_grade_encoder)
X_test_cat['cb_person_default_on_file'] = X_test_cat['cb_person_default_on_file'].map(cb_person_default_on_file_encoder)
```

```python
X_test_transformed = pd.concat([X_test_num_transformed,X_test_cat], axis = 1)
```

```python
X_test_transformed.head()
```

## logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

classifier = LogisticRegression()
classifier.fit(X_train_transformed, y_train)

y_test_pred = classifier.predict(X_test_transformed)

print(accuracy_score(y_test, y_test_pred))

print(classification_report(y_test, y_test_pred))
```

```python
output_df= pd.DataFrame({'Actual': y_test.values.flatten()}, index=X_test.index)
```

```python
output_df['Logistic Regression Predictions'] = y_test_pred

output_df
```

## DT Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report


classifier = DecisionTreeClassifier()
classifier.fit(X_train_transformed, y_train)
y_test_pred = classifier.predict(X_test_transformed)

print(accuracy_score(y_test, y_test_pred))

print(classification_report(y_test, y_test_pred))
```

```python
output_df['DT classifier Predictions'] = y_test_pred

output_df
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report


classifier = RandomForestClassifier()
classifier.fit(X_train_transformed, y_train)
y_test_pred = classifier.predict(X_test_transformed)

print(accuracy_score(y_test, y_test_pred))

print(classification_report(y_test, y_test_pred))
```

```python
output_df[' randomforest Predictions'] = y_test_pred

output_df
```

## Naive bayes

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn import metrics

classifier = BernoulliNB()
classifier.fit(X_train_transformed, y_train)

y_test_pred = classifier.predict(X_test_transformed)

print(metrics.accuracy_score(y_test, y_test_pred))
```

```python
output_df['naive bayes'] = y_test_pred

output_df
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier()
classifier.fit(X_train_transformed, y_train)
y_test_pred = classifier.predict(X_test_transformed)
print(accuracy_score(y_test, y_test_pred))

print(classification_report(y_test, y_test_pred))
```

```python
output_df[' KNN classifier Predictions'] = y_test_pred

output_df
```
